Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.
- Answer:

Anomaly Detection is the process of identifying data points or patterns that are different from normal data behavior. These unusual data points are called anomalies or outliers. Anomaly detection is mainly used to detect fraud, system failures, unusual activities, or errors in data.

There are three main types of anomalies:

Point Anomaly:
A point anomaly occurs when a single data value is very different from the rest of the data.
Example: If most customers spend around ₹2,000 and one customer suddenly spends ₹1,00,000, it is a point anomaly.

Contextual Anomaly:
A contextual anomaly depends on a specific situation such as time or condition.
Example: High electricity usage is normal during summer but the same usage during winter can be considered abnormal.

Collective Anomaly:
A collective anomaly occurs when a group of data points together behaves abnormally.
Example: A continuous drop in server performance for several hours may indicate a system issue.

**Question 2:** Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.
- Answer:

Isolation Forest identifies anomalies by randomly separating data points. Anomalies are isolated faster because they are different from normal data. It works well with large and high-dimensional datasets.

DBSCAN is a density-based clustering algorithm. It forms clusters of dense data points and treats points outside clusters as anomalies. It is useful when the data has clusters and noise.

Local Outlier Factor (LOF) detects anomalies by comparing the density of a data point with its neighboring points. If a point has much lower density than its neighbors, it is treated as an outlier. It is useful for detecting local anomalies.

**Question 3 :** What are the key components of a Time Series? Explain each with one example.
- Answer:

A time series is a set of observations collected over time. The key components of a time series are:

Trend:
It represents the long-term movement in the data.
Example: Increase in yearly internet users.

Seasonality:
It shows repeating patterns at regular intervals.
Example: Higher sales of sweaters during winter.

Cyclic Component:
These are patterns that repeat over time but not at fixed intervals.
Example: Economic growth and recession cycles.

Residual (Noise):
Random variations that cannot be explained by trend or seasonality.
Example: Sudden power failure affecting electricity usage.

**Question 4:** Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?
- Answer:

A time series is called stationary if its mean, variance, and overall pattern remain constant over time. Stationary data is important because most time series models work better on stationary data.

Stationarity can be tested using the Augmented Dickey-Fuller (ADF) test. If the p-value is less than 0.05, the series is stationary.

A non-stationary series can be transformed into stationary by:

Applying differencing

Using logarithmic transformation

Removing trend and seasonality

**Question 5:** Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.
- Answer:

AR (AutoRegressive): Uses past values of the time series to predict future values.




MA (Moving Average): Uses past forecast errors to make predictions.

SARIMA: An extension of ARIMA that also handles seasonal patterns.

SARIMAX: SARIMA model that includes external variables like weather or holidays.

**Question 7:** Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to detect anomalies. Visualize the anomalies on a 2D scatter plot. (Include your Python code and output in the code box below.)
- **Answer:**

```python
from sklearn.ensemble import IsolationForest
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(ts, columns=["Passengers"])

model = IsolationForest(contamination=0.05, random_state=1)
df["Anomaly"] = model.fit_predict(df)

plt.figure(figsize=(8,4))
plt.scatter(df.index, df["Passengers"], c=df["Anomaly"])
plt.show()
```

**Question 8:** Train a SARIMA model on the monthly airline passengers dataset. Forecast the next 12 months and visualize the results. (Include your Python code and output in the code box below.)
- **Answer:**

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt

model = SARIMAX(ts, order=(1,1,1), seasonal_order=(1,1,1,12))
result = model.fit()

forecast = result.forecast(12)

plt.figure(figsize=(8,4))
plt.plot(ts)
plt.plot(forecast)
plt.show()
```

**Question 10:** You are working as a data scientist for a power grid monitoring company.
Your goal is to forecast energy demand and also detect abnormal spikes or drops in
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.
Explain your real-time data science workflow:
● How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?
● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?
● How would you validate and monitor the performance over time?
● How would this solution help business decisions or operations?
(Include your Python code and output in the code box below.)

**Answer** - In a power grid monitoring system, data is received continuously every 15 minutes. The first step is to clean the data and handle missing values so that incorrect readings do not affect the system.

For anomaly detection, Isolation Forest or LOF can be used because they work well with numerical data and can quickly identify unusual spikes or sudden drops in energy usage. These methods help in detecting abnormal behavior in real time.

For short-term forecasting, SARIMAX is suitable because energy demand depends not only on past usage but also on external factors like weather and region. This model helps in predicting near-future demand accurately.

Model performance can be checked using error metrics like MAE or RMSE and by comparing predicted values with actual consumption regularly. The model should be retrained from time to time to maintain accuracy.

This solution helps the company by detecting faults early, preventing power failures, managing load efficiently, and improving decision-making for power distribution.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from statsmodels.tsa.statespace.sarimax import SARIMAX

np.random.seed(1)
dates = pd.date_range("2024-01-01", periods=200, freq="15min")
usage = np.random.normal(500, 50, 200)
usage[50] = 900
usage[120] = 100

df = pd.DataFrame({"usage": usage}, index=dates)

model_if = IsolationForest(contamination=0.05)
df["anomaly"] = model_if.fit_predict(df[["usage"]])

plt.scatter(df.index, df["usage"], c=df["anomaly"])
plt.show()

model = SARIMAX(df["usage"], order=(1,1,1))
result = model.fit()

forecast = result.forecast(20)

plt.plot(df["usage"])
plt.plot(forecast)
plt.show()
```
